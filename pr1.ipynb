```python
import pandas as pd
import numpy as nm
```

```python
dataframe = {
    "Name": [
        'Shivansh', 'Neekil', 'Shirish', 'Yashwant', 'Omkar', 
        'Atharva', 'Anita', 'Raghvendra', 'Tarun', 'Vikram', 
        'Pooja', 'Amit', 'Neha', 'Tanya', 'Divya', 
        'Arjun', 'Kavya', 'Rajni', 'Meera', 'Nitin'
    ],
    "Age": [
        18, 19, 20, 21, 22,
        23, 19, 24, 20, 25,
        21, 22, 23, 24, 19,
        22, 20, 23, 21, 24
    ],
    "Gender": [
        'M', 'M', 'M', 'M', 'M',
        'M', 'F', 'M', 'M', 'M',
        'F', 'M', 'F', 'F', 'F',
        'M', 'F', 'F', 'F', 'M'
    ],
    "Marks": [
        80, nm.nan, 90, 85, nm.nan,
        88, 92, nm.nan, 86, 79,
        91, nm.nan, 84, 87, 93,
        82, nm.nan, 89, 85, nm.nan
    ]
}

df = pd.DataFrame(dataframe)
print('Original Table', end = ' ')
df
```

```python
print('\nMissing Values: ')
df.isnull()
```

```python
cat = []
con = []

for i in df.columns:
    if (df[i].dtype == "object"):
        cat.append(i)
    else:
        con.append(i)
```

```python
print('Categorical Data is: ', end = ' ')
print(cat, end = '\n')
print('Continuous Data: ', end = ' ')
print(con)
```

```python
c = avg = total_marks = 0

for ele in df['Marks']:
    if pd.notnull(ele):
        c += 1
        total_marks += ele

if c > 0:
    avg = total_marks / c
else:
    avg = 0

print('\nAverage Marks: ', avg)
```

```python
df['Marks'] = df['Marks'].fillna(avg)
print('Updated Table: ', end = ' ')
df
```

```python
df_copy = df.copy()
df_copy = df_copy.drop(['Age'], axis = 1)
print('\nAfter Dropping Age Column: ')
df_copy
```

```python
df_temp = df.copy()
print('\nStudents with Marks > Average: ')
df_temp = df[df['Marks'] > avg]
df_temp
```

```python
df_copy['Gender'] = df_copy['Gender'].map({'M': 0, 'F': 1}).astype(int)
print('\nFinal DataFrame: ')
df_copy
```
